### 1. Arithmetic Operators (ตัวดำเนินการทางคณิตศาสตร์)  
ใช้สำหรับคำนวณค่าทางคณิตศาสตร์ เช่น บวก ลบ คูณ หาร และ Modulo  

#### ตัวอย่างโค้ด:  
```python  
# ตัวอย่างการคำนวณราคาอาหาร  
price_pizza = 100  
price_drink = 35  
total = price_pizza + price_drink  # บวก (Addition)  
print("Total:", total)  # 135  

# คำนวณส่วนลด 10%  
discount = total * 0.10  # คูณ (Multiplication)  
new_total = total - discount  # ลบ (Subtraction)  
print("After discount:", new_total)  # 121.5  

# หารหาจำนวนเงินทอน  
payment = 150  
change = payment - new_total  # 150 - 121.5 = 28.5  
print("Change:", change)  # 28.5  

# Modulo (หาเศษ)  
print(10 % 3)  # Output: 1 (10 หารด้วย 3 เหลือเศษ 1)  
```  

#### ข้อความ Thai:  
- **บวก (+):** ใช้รวมราคาอาหารและเครื่องดื่ม  
- **ลบ (-):** ใช้หักส่วนลดหรือคำนวณเงินทอน  
- **คูณ (*):** ใช้คำนวณส่วนลดหรือราคาอาหารหลายชิ้น  
- **หาร (/):** ใช้หารจำนวนคนหรือคำนวณราคาเฉลี่ย  
- **Modulo (%):** ใช้หาเศษเมื่อหาร เช่น ตรวจสอบว่าจำนวนโต๊ะเป็นเลขคู่หรือคี่  

---

### 2. Comparison Operators (ตัวดำเนินการเปรียบเทียบ)  
ใช้เปรียบเทียบค่า 2 ค่า เพื่อตรวจสอบเงื่อนไข เช่น ตรวจสอบว่าราคาอาหารเกิน 500 บาทหรือไม่  
#### สัญลักษณ์ในการเปรียบเทียบ:  
- **== (เท่ากับ):** เช็คยอดขายวันนี้เท่ากับเป้าหมายหรือไม่  
- **!= (ไม่เท่ากับ):** เช็ค ID โต๊ะซ้ำหรือไม่  
- **> (มากกว่า), < (น้อยกว่า):** เช็คราคาอาหารเกินหรือไม่  
- **>=, <=:** เช็คยอดเงินทอนเพียงพอหรือไม่ 
#### ตัวอย่างโค้ด:  
```python  
price = 450  
minimum_spend = 500  

if price >= minimum_spend:  
    print("คุณได้รับส่วนลดพิเศษ!")  
else:  
    print("สั่งเพิ่มได้อีก", minimum_spend - price, "บาทเพื่อรับส่วนลด")  

# Output: สั่งเพิ่มได้อีก 50 บาทเพื่อรับส่วนลด  
```

### 3. Logical Operators (ตัวดำเนินการตรรกะ)  
ใช้เชื่อมเงื่อนไขหลายเงื่อนไข เช่น ลูกค้าต้องเป็นสมาชิก **และ** ยอดซื้อเกิน 1,000 บาท  
#### ข้อความ Thai:  
- **and:** เชื่อมเงื่อนไขที่ต้องเป็นจริงทั้งสอง (เช่น สมาชิก + ยอดซื้อ)  
- **or:** เชื่อมเงื่อนไขที่เป็นจริงอย่างน้อยหนึ่ง (เช่น เริ่ม shift ตอนเช้า หรือ บ่าย)  
- **not:** กลับค่าเงื่อนไข (เช่น ไม่ใช่เวลาปิดร้าน) 

#### ตัวอย่างโค้ด:  
```python  
is_member = True  
total = 1200  
minimum_for_discount = 1000  

if is_member and total >= minimum_for_discount:  
    print("คุณได้รับส่วนลดสมาชิก!")  
else:  
    print("ไม่ได้รับส่วนลด")  

# Output: คุณได้รับส่วนลดสมาชิก!  
```

### 4. Assignment Operators (ตัวดำเนินการกำหนดค่า)  
ใช้กำหนดค่าให้ตัวแปรอย่างรวดเร็ว โดยรวมกับการคำนวณ  
#### ข้อความ Thai:  
- **+=, -=, *=, /=, %=:** ช่วยลดขั้นตอนการคำนวณและกำหนดค่าใหม่  

#### ตัวอย่างโค้ด:  
```python  
# ตัวอย่างการอัปเดตยอดขาย  
sales = 500  
sales += 200  # เท่ากับ sales = sales + 200 → 700  
sales -= 50    # 650  
sales *= 2     # 1300  
sales /= 10    # 130.0  
```

### 5. Bitwise Operators (ตัวดำเนินการบิต)  
ทำงานกับค่าบิตของตัวเลข ใช้ในสถานการณ์เฉพาะ เช่น ตั้งค่าสถานะ  

#### ข้อความ Thai:  
- **& (และบิต), | (หรือบิต), ^ (XOR), << (เลื่อนซ้าย):** ใช้จัดการสถานะหลายอย่างในตัวแปรเดียว  

#### ตัวอย่างโค้ด:  
```python  
# Assign shifts using bitwise (Mon=1, Tue=2, Wed=4, Thu=8, Fri=16)  
chef01_shifts = 0b10110  # Works Mon, Wed, Thu 
requested_day = 0b10000  # Friday 


if chef_shifts & requested_day:  
    print("Chef is available Friday!")
```

### 6. Order of Precedence (ลำดับการคำนวณ)  
**Purpose**: Control the order of operations.
**Rule**: Parentheses → Exponents → Multiplication/Division → Addition/Subtraction → Bitwise → Logical.
ตัวดำเนินการที่มีลำดับสูงสุดถูกคำนวณก่อน:  
1. **(), \*\***  
2. **\*, /, //, %, +, -**
3. **<<, >>**  
4. **&, ^, |**  
5. **==, !=, >, <, >=, <=**  
6. **not**  
7. **and**  
8. **or**  

#### คำแนะนำ:  
- เพื่อป้องกันความสับสน ควรใส่วงเล็บเพื่อกำหนดลำดับการคำนวณให้ชัดเจน
