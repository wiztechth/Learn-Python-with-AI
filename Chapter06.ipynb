**คำอธิบายเนื้อหา (Explanation of Content)**

ในบทนี้ เราจะมาเรียนรู้การทำให้โปรแกรมของเราสามารถ "ตัดสินใจ" ได้ โดยการใช้ **คำสั่งเงื่อนไข (Conditional Statements)** ครับ นี่เป็นหัวใจสำคัญของการเขียนโปรแกรม เพราะชีวิตจริงและการทำงานของร้านอาหารก็เต็มไปด้วยการตัดสินใจตามเงื่อนไขต่างๆ เช่น ถ้าโต๊ะว่าง ก็นำลูกค้าไปนั่ง ถ้าออเดอร์ถึงยอดที่กำหนด ก็ให้ส่วนลด ถ้าเป็นช่วง Happy Hour ก็ลดราคาเครื่องดื่ม เป็นต้น คำสั่งเงื่อนไขจะช่วยให้โปรแกรมของเราตอบสนองต่อสถานการณ์ที่แตกต่างกันได้อย่างเหมาะสม

**ทำความเข้าใจตรรกะแบบมีเงื่อนไข (Understanding Conditional Logic)**

ตรรกะแบบมีเงื่อนไขคือแนวคิดที่ว่า "ถ้าเงื่อนไขนี้เป็นจริง ให้ทำสิ่งนี้" และอาจจะมีส่วนเสริมว่า "มิฉะนั้น (ถ้าเงื่อนไขไม่จริง) ให้ทำอีกสิ่งหนึ่ง" นี่คือพื้นฐานที่เราจะนำไปสร้างคำสั่งต่างๆ ครับ

**คำสั่ง if (if Statements)**

เป็นคำสั่งเงื่อนไขที่พื้นฐานที่สุด ใช้เพื่อตรวจสอบเงื่อนไขเดียว ถ้าเงื่อนไขเป็นจริง โปรแกรมจะทำตามชุดคำสั่งที่อยู่ภายใต้ `if` แต่ถ้าไม่จริง โปรแกรมจะข้ามส่วนนี้ไป

**คำสั่ง if-else (if-else Statements)**

คำสั่งนี้เพิ่มทางเลือกเข้ามา ถ้าเงื่อนไขเป็นจริง ให้ทำชุดคำสั่งหนึ่ง แต่ถ้าเงื่อนไขเป็น *ไม่จริง* (เข้าสู่ส่วน `else`) ให้ไปทำอีกชุดคำสั่งหนึ่งแทน มีสองทางเลือกเสมอ

**โครงสร้าง if-elif-else แบบขั้นบันได (if-elif-else Ladder)**

เมื่อเราต้องการตรวจสอบหลายๆ เงื่อนไขตามลำดับ เราจะใช้โครงสร้าง `if-elif-else` (`elif` ย่อมาจาก "else if") โปรแกรมจะตรวจสอบเงื่อนไขแรก (`if`) ถ้าไม่จริง จะไปตรวจสอบเงื่อนไขที่สอง (`elif`) ถ้ายังไม่จริง ก็ตรวจสอบ `elif` ถัดไปเรื่อยๆ จนกว่าจะเจอเงื่อนไขที่เป็นจริง แล้วทำชุดคำสั่งภายใต้เงื่อนไขนั้น จากนั้นก็จะออกจากโครงสร้างทั้งหมด ถ้าไม่มีเงื่อนไขใดเป็นจริงเลย โปรแกรมจะทำชุดคำสั่งในส่วน `else` (ถ้ามี)

**คำสั่งเงื่อนไขแบบซ้อนกัน (Nested Conditional Statements)**

เราสามารถวางคำสั่งเงื่อนไข (`if`, `if-else`, `if-elif-else`) ไว้ *ภายใน* คำสั่งเงื่อนไขอื่นได้ เรียกว่า Nested Condition ใช้เมื่อการตัดสินใจต้องอาศัยการตรวจสอบเงื่อนไขย่อยๆ หลังจากเงื่อนไขหลักเป็นจริงแล้ว เช่น ถ้าโต๊ะว่าง (เงื่อนไขหลัก) *แล้วค่อย* ตรวจสอบต่อว่าจำนวนลูกค้าเหมาะสมกับขนาดโต๊ะหรือไม่ (เงื่อนไขซ้อน)

**ตัวอย่างเชิงปฏิบัติในธุรกิจร้านอาหาร (Practical Examples in Restaurant Business)**

เราจะใช้ตัวอย่างต่างๆ ที่เกี่ยวข้องกับการทำงานในร้านอาหาร เช่น การจัดการโต๊ะ การคิดเงิน การให้ส่วนลด การจัดการสต็อกสินค้า เพื่อแสดงให้เห็นว่าคำสั่งเงื่อนไขมีประโยชน์อย่างไร

**Sample Code (ตัวอย่างโค้ด)**

ต่อไปนี้คือตัวอย่างโค้ดภาษา Python ที่แสดงการใช้คำสั่งเงื่อนไขในสถานการณ์ร้านอาหาร (โค้ดเป็นภาษาอังกฤษตามที่ระบุ)

**1. if Statement: Checking Table Availability (ตรวจสอบว่าโต๊ะว่างหรือไม่)**

```python
# Assume we have a variable representing table status
table_status = "occupied" # Could be "free", "occupied", "reserved"

# Check if the table is free
if table_status == "free":
    print("Table is available. You can seat the customer.")

# If table_status is "occupied" or "reserved", nothing will be printed by this if statement.
```
*คำอธิบาย:* โค้ดนี้จะตรวจสอบว่าสถานะของโต๊ะคือ "free" หรือไม่ ถ้าใช่ จะพิมพ์ข้อความว่าโต๊ะว่าง
---

**2. if-else Statement: Applying Member Discount (ให้ส่วนลดสมาชิกหรือไม่)**

```python
# Assume variables for total bill and member status
bill_total = 1500
is_member = True # True if the customer is a member, False otherwise

# Check if the customer is a member to apply a discount
if is_member:
    discount_rate = 0.10 # 10% discount
    discount_amount = bill_total * discount_rate
    final_bill = bill_total - discount_amount
    print(f"Customer is a member. Applied {discount_rate*100}% discount.")
    print(f"Final bill: {final_bill} Baht")
else:
    print("Customer is not a member. No discount applied.")
    print(f"Final bill: {bill_total} Baht")
```
*คำอธิบาย:* โค้ดนี้ตรวจสอบว่าลูกค้าเป็นสมาชิกหรือไม่ ถ้าเป็น (`True`) จะคำนวณส่วนลด 10% ถ้าไม่เป็น (`False`) จะไม่มีส่วนลด

---
**3. if-elif-else Ladder: Determining Discount Tier by Bill Total (กำหนดระดับส่วนลดตามยอดบิล)**

```python
# Assume a variable for the total bill amount
bill_total = 3500

# Determine discount tier based on bill total
if bill_total >= 5000:
    discount_percentage = 20
    print("Bill total >= 5000. Applied 20% discount.")
elif bill_total >= 2000:
    discount_percentage = 10
    print("Bill total >= 2000. Applied 10% discount.")
elif bill_total >= 1000:
    discount_percentage = 5
    print("Bill total >= 1000. Applied 5% discount.")
else:
    discount_percentage = 0
    print("Bill total < 1000. No discount applied.")

discount_amount = bill_total * (discount_percentage / 100)
final_bill = bill_total - discount_amount
print(f"Original bill: {bill_total} Baht")
print(f"Discount applied: {discount_amount} Baht")
print(f"Final bill: {final_bill} Baht")
```
*คำอธิบาย:* โค้ดนี้จะตรวจสอบยอดรวมบิลตามลำดับ เพื่อกำหนดเปอร์เซ็นต์ส่วนลดที่แตกต่างกัน ถ้าไม่เข้าเงื่อนไขใดเลย ก็จะไม่มีส่วนลด

---
**4. Nested Conditional Statements: Seating Customer (จัดที่นั่งให้ลูกค้า)**

```python
# Assume variables for table status and customer count
table_status = "free"
table_size = 4 # Number of seats at the table
customer_count = 3

# Check if table is free, THEN check if size is appropriate
if table_status == "free":
    print(f"Table (size {table_size}) is available.")
    # Nested condition: Check if the table size is suitable for the customer count
    if customer_count <= table_size:
        print(f"Table size {table_size} is suitable for {customer_count} customers.")
        print("Proceed to seat the customer.")
    else:
        print(f"Table size {table_size} is too small for {customer_count} customers.")
        print("Look for another table.")
else:
    print(f"Table is {table_status}. Cannot seat customer.")
```
*คำอธิบาย:* โค้ดนี้จะตรวจสอบก่อนว่าโต๊ะว่างหรือไม่ (`if table_status == "free"`) ถ้าว่าง จึงจะเข้าสู่การตรวจสอบเงื่อนไขที่ซ้อนอยู่ (`if customer_count <= table_size`) เพื่อดูว่าขนาดโต๊ะพอสำหรับจำนวนลูกค้าหรือไม่
