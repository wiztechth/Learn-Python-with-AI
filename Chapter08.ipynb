### **นาทีที่ 0:45 - 2:00 | 1. Introduction to Dictionaries**
**(เสียงผู้ชาย)** มาเริ่มกันเลยครับ Dictionary ใน Python เนี่ย จะถูกสร้างขึ้นมาโดยใช้วงเล็บปีกกา `{}` ครับ และภายในจะประกอบด้วยคู่ของ `Key` และ `Value` คั่นด้วยเครื่องหมายทวิภาค `:` และแต่ละคู่ Key-Value จะคั่นกันด้วยเครื่องหมายจุลภาค `,` ครับ

**(บนหน้าจอ: โค้ดสร้าง Dictionary เมนูอาหาร)**
```python
# ตัวอย่าง Dictionary เมนูอาหาร (ชื่อเมนูเป็น Key, ราคาเป็น Value)
menu = {
    "ข้าวผัดกะเพรา": 70,
    "ต้มยำกุ้ง": 180,
    "ผัดไทยกุ้งสด": 90,
    "แกงเขียวหวานไก่": 150
}

print("ประเภทของตัวแปร menu คือ:", type(menu))
print("เมนูทั้งหมด:", menu)
```
**(เสียงผู้ชาย)** เห็นไหมครับ ชื่อเมนูเป็น Key ซึ่งต้องไม่ซ้ำกัน ส่วนราคาเป็น Value ครับ Key ส่วนใหญ่มักจะเป็น String หรือตัวเลข แต่ต้องเป็นข้อมูลประเภทที่เปลี่ยนแปลงไม่ได้ (Immutable) นะครับ ส่วน Value เนี่ย เป็นข้อมูลประเภทอะไรก็ได้เลยครับ

**(บนหน้าจอ: โค้ดเข้าถึง Value)**
```python
# การเข้าถึงราคาของเมนูด้วยชื่อ (Key)
price_kaprao = menu["ข้าวผัดกะเพรา"]
print("ราคาข้าวผัดกะเพรา:", price_kaprao, "บาท")
```
**(เสียงผู้ชาย)** การเข้าถึง Value ทำได้ง่ายๆ โดยใช้ Key ในวงเล็บเหลี่ยม `[]` แบบนี้ครับ แต่ถ้า Key ที่เราพยายามเข้าถึงไม่มีอยู่ใน Dictionary ล่ะ? Python ก็จะแจ้งข้อผิดพลาดที่เรียกว่า `KeyError` ครับ

**(บนหน้าจอ: โค้ดแสดง KeyError และการใช้ .get())**
```python
# การเข้าถึง Key ที่ไม่มีอยู่จะทำให้เกิดข้อผิดพลาด KeyError
# price_somtum = menu["ส้มตำ"] # uncommenting this line will cause an error

# การใช้ .get() เพื่อเข้าถึงค่าอย่างปลอดภัย
# ถ้า Key ไม่มี จะคืนค่า None หรือค่า Default ที่เรากำหนด
price_somtum_safe = menu.get("ส้มตำ", "ไม่พบเมนูนี้")
print("ราคาตำส้มตำ (แบบปลอดภัย):", price_somtum_safe)

price_tom_yum_safe = menu.get("ต้มยำกุ้ง", "ไม่พบเมนูนี้")
print("ราคาต้มยำกุ้ง (แบบปลอดภัย):", price_tom_yum_safe, "บาท")
```
**(เสียงผู้ชาย)** ดังนั้น การใช้เมธอด `.get()` จึงเป็นวิธีที่ปลอดภัยกว่าในการเข้าถึง Value ครับ

### **นาทีที่ 2:00 - 3:30 | 2. Adding, Updating, and Deleting Key-Value Pairs**
**(เสียงผู้ชาย)** ต่อมา เรามาดูวิธีการจัดการข้อมูลใน Dictionary กันบ้างครับ ไม่ว่าจะเป็นการเพิ่มรายการใหม่ อัพเดทข้อมูล หรือลบรายการที่ไม่ต้องการแล้ว

**(บนหน้าจอ: โค้ดเพิ่มและอัพเดท)**
```python
# สร้าง Dictionary คลังสินค้า (ชื่อสินค้าเป็น Key, จำนวนเป็น Value)
inventory = {
    "ข้าวสาร": 50, # หน่วยเป็น กก.
    "เส้นใหญ่": 20, # หน่วยเป็น กก.
    "เนื้อไก่": 30, # หน่วยเป็น กก.
    "กุ้งสด": 15   # หน่วยเป็น กก.
}
print("\nคลังสินค้าก่อนการเปลี่ยนแปลง:", inventory)

# เพิ่มรายการใหม่ (ไข่ไก่)
inventory["ไข่ไก่"] = 100 # หน่วยเป็น ฟอง
print("คลังสินค้าหลังจากเพิ่ม 'ไข่ไก่':", inventory)

# อัพเดทจำนวนสินค้า (ข้าวสารใช้ไป 5 กก.)
inventory["ข้าวสาร"] = inventory["ข้าวสาร"] - 5
print("คลังสินค้าหลังจากอัพเดท 'ข้าวสาร':", inventory)
```
**(เสียงผู้ชาย)** การเพิ่มหรืออัพเดททำได้โดยใช้ Key ในวงเล็บเหลี่ยม `[]` เหมือนเดิมครับ ถ้า Key นั้นยังไม่มีใน Dictionary ก็จะเป็นการ **เพิ่ม** คู่ Key-Value ใหม่เข้าไป แต่ถ้า Key นั้นมีอยู่แล้ว ก็จะเป็นการ **อัพเดท** Value ของ Key นั้นครับ

**(บนหน้าจอ: โค้ดลบด้วย del และ pop)**
```python
# ลบรายการโดยใช้ del (เส้นใหญ่หมด)
del inventory["เส้นใหญ่"]
print("คลังสินค้าหลังจากลบ 'เส้นใหญ่' ด้วย del:", inventory)

# ลบรายการโดยใช้ pop() และเก็บค่าที่ลบ (กุ้งสดหมด)
removed_quantity = inventory.pop("กุ้งสด")
print("ปริมาณกุ้งสดที่ถูกลบออก:", removed_quantity, "กก.")
print("คลังสินค้าหลังจากลบ 'กุ้งสด' ด้วย pop():", inventory)

# pop() สามารถกำหนดค่า default ได้ ถ้า Key ไม่มีก็จะไม่ error
# inventory.pop("ผักชี", "ไม่มีผักชีให้ลบ")
```
**(เสียงผู้ชาย)** สำหรับการลบ เรามีสองวิธีหลักๆ ครับ คือใช้คำสั่ง `del` ตามด้วยชื่อ Dictionary และ Key ที่ต้องการลบ หรือใช้เมธอด `.pop(key)` ซึ่ง `.pop()` จะคืนค่า Value ของ Key ที่ถูกลบออกมาก่อนลบครับ

### **นาทีที่ 3:30 - 5:00 | 3. Dictionary Methods: keys(), values(), items()**
**(เสียงผู้ชาย)** Dictionary ยังมีเมธอด (Method) ที่มีประโยชน์มากๆ อีกหลายตัวครับ ที่นิยมใช้กันบ่อยๆ ก็คือ `.keys()`, `.values()`, และ `.items()` ครับ

**(บนหน้าจอ: โค้ดแสดงผล keys(), values(), items())**
```python
# ใช้ Dictionary คลังสินค้า
inventory = {
    "ข้าวสาร": 45,
    "เนื้อไก่": 30,
    "ไข่ไก่": 100
}

print("\nรายการสินค้าทั้งหมด (Keys):", inventory.keys())
print("รายการปริมาณสินค้าทั้งหมด (Values):", inventory.values())
print("รายการสินค้าและปริมาณ (Items):", inventory.items())
```
**(เสียงผู้ชาย)** เมธอดเหล่านี้จะคืนค่าเป็น View Object ซึ่งเป็นเหมือนมุมมองที่เชื่อมโยงกับ Dictionary ต้นฉบับครับ `.keys()` จะให้เฉพาะ Key, `.values()` จะให้เฉพาะ Value, ส่วน `.items()` จะให้เป็นคู่ Key-Value ในรูปแบบของ Tuple ครับ ซึ่ง View Object พวกนี้มีประโยชน์มากเวลาเราต้องการวนลูป (Loop) ผ่าน Dictionary ครับ

### **นาทีที่ 5:00 - 6:30 | 4. Iterating Over Dictionaries**
**(เสียงผู้ชาย)** การวนลูปหรือ Iterate ผ่าน Dictionary เป็นสิ่งที่เราทำบ่อยมากครับ เราสามารถวนลูปเพื่อเข้าถึง Key, Value หรือทั้งคู่พร้อมกันได้

**(บนหน้าจอ: โค้ดวนลูปแบบต่างๆ)**
```python
# ใช้ Dictionary เมนูอาหารอีกครั้ง
menu = {
    "ข้าวผัดกะเพรา": 70,
    "ต้มยำกุ้ง": 180,
    "ผัดไทยกุ้งสด": 90,
    "แกงเขียวหวานไก่": 150
}

print("\n--- วนลูปผ่าน Key (ชื่อเมนู) ---")
for dish_name in menu: # นี่คือการวนลูปผ่าน Key โดยตรง
    print(dish_name)

print("\n--- วนลูปผ่าน Value (ราคา) ---")
for price in menu.values():
    print(price, "บาท")

print("\n--- วนลูปผ่านทั้ง Key และ Value (รายการเมนูและราคา) ---")
for dish_name, price in menu.items():
    print(f"เมนู: {dish_name}, ราคา: {price} บาท")
```
**(เสียงผู้ชาย)** เห็นไหมครับ การวนลูปผ่าน `.items()` เป็นวิธีที่สะดวกและนิยมที่สุดเมื่อเราต้องการใช้ทั้ง Key และ Value ในแต่ละรอบของการวนลูปครับ ลองนำไปประยุกต์ใช้ในการคำนวณยอดรวมของออเดอร์ลูกค้าดูสิครับ

**(บนหน้าจอ: โค้ดคำนวณยอดรวมออเดอร์)**
```python
# ตัวอย่างการใช้ในการคำนวณยอดรวม (สมมติมีรายการที่ลูกค้าสั่ง)
order = {"ข้าวผัดกะเพรา": 2, "ผัดไทยกุ้งสด": 1, "ต้มยำกุ้ง": 1}
total_bill = 0
print("\n--- รายการที่ลูกค้าสั่งและคำนวณยอดรวม ---")
for item, quantity in order.items():
    item_price = menu.get(item, 0) # ใช้ .get() เพื่อป้องกัน error ถ้าเมนูไม่อยู่ใน menu หลัก
    item_total = item_price * quantity
    print(f"{item} x {quantity} = {item_total} บาท")
    total_bill += item_total

print("ยอดรวมทั้งหมด:", total_bill, "บาท")
```

### **นาทีที่ 6:30 - 7:30 | 5. Dictionary Comprehensions**
**(เสียงผู้ชาย)** สำหรับคนที่อยากเขียนโค้ดให้กระชับและอ่านง่ายขึ้น Python ก็มีสิ่งที่เราเรียกว่า **Dictionary Comprehension** ครับ มันเป็นวิธีสร้าง Dictionary ใหม่จากโครงสร้างข้อมูลอื่น หรือจาก Dictionary เดิมของเรา ด้วยโค้ดแค่บรรทัดเดียวครับ

**(บนหน้าจอ: โค้ด Dictionary Comprehension)**
```python
# ใช้ Dictionary เมนูอาหาร
menu = {
    "ข้าวผัดกะเพรา": 70,
    "ต้มยำกุ้ง": 180,
    "ผัดไทยกุ้งสด": 90,
    "แกงเขียวหวานไก่": 150
}

print("\nเมนูเดิม:", menu)

# สร้าง Dictionary ราคาใหม่ที่บวกภาษี 7% ด้วย Dictionary Comprehension
menu_with_tax = {dish: price * 1.07 for dish, price in menu.items()}
print("เมนูพร้อมราคาที่บวกภาษี 7%:", menu_with_tax)

# สร้าง Dictionary เฉพาะเมนูที่มีราคาต่ำกว่า 100 บาท
affordable_menu = {dish: price for dish, price in menu.items() if price < 100}
print("เมนูราคาไม่เกิน 100 บาท:", affordable_menu)
```
**(เสียงผู้ชาย)** สะดวกและรวดเร็วใช่ไหมครับ Dictionary Comprehension ช่วยให้โค้ดของเราดูสะอาดตามากขึ้นเยอะเลยครับ
